### 4.Highest and Lowest total score in an innings

```python
total_score = df.groupby(['match_id','inning']).total_runs.sum()
```

```python
total_score
```

```python
total_score.head(1)
```

```python
total_score.max()
```

```python
total_score.min()
```

```python
# Technique 2
total_score1 = df.groupby(['match_id','inning']).total_runs.sum().sort_values(ascending=False)
```

```python
print('Highest')
print(total_score1.head(1))
```

```python
print('Lowest')
print(total_score1.tail(1))
```

```python
# Technique 3
## Reset index() function is used to generate a new DataFrame or Series with the index reset.
## This is usefull when index needs to be treated as column ,or when a index is meaningless and
## needs to be reset to the default before another operation.
total_score2 = total_score.reset_index() 
```

```python
total_score2
```

```python
print('Highest',total_score2['total_runs'].max())
```

```python
print('Lowest',total_score2['total_runs'].min())
```
